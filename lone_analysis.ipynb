# Part B – Loan Default Analysis
This notebook performs statistical and probability analysis on loan data. Each step includes explanation of what is done and why it is required.

## Step 1 – Import Libraries
**What is done:** Required Python libraries are imported.

**Why:** These libraries help in data analysis, visualization and statistical calculations.
- pandas → data handling
- numpy → numerical operations
- matplotlib & seaborn → graphs
- scipy → statistical plots

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Step 2 – Load Dataset
**What is done:** Dataset file `loan_applications.csv` is loaded.

**Why:** Data must be loaded before performing statistical operations.
`head()` shows first 5 records to verify dataset is loaded correctly.

```
df = pd.read_csv('loan_applications.csv')
df.head()
```

## Step 3 – Central Tendency (Mean, Median, Mode)
**What is done:** Income column is analyzed using Mean, Median and Mode.

**Why:** These values help understand the typical income of customers.

```
print('Mean Income:', df['Income'].mean())
print('Median Income:', df['Income'].median())
print('Mode Income:', df['Income'].mode()[0])
```

## Step 4 – Dispersion Measures
**What is done:** Range, Variance and Standard Deviation of Loan Amount are calculated.

**Why:** These measures help understand how loan amounts vary from average value.

```
loan_range = df['Loan_Amount'].max() - df['Loan_Amount'].min()
print('Range:', loan_range)
print('Variance:', df['Loan_Amount'].var())
print('Standard Deviation:', df['Loan_Amount'].std())
```

## Step 5 – Probability of Default
**What is done:** Default probability is calculated.

**Why:** This helps understand risk level of customers.

```
default_probability = (df['Default_Status']=='Yes').mean()
print('Probability of Default:', default_probability)
```

## Step 6 – Contingency Table
**What is done:** Credit scores are grouped into categories and compared with default status.

**Why:** This helps identify how credit score affects default behavior.

```
df['Credit_Category']=pd.cut(df['Credit_Score'],bins=[300,600,700,850],labels=['Low','Medium','High'])
pd.crosstab(df['Default_Status'],df['Credit_Category'])
```

## Step 7 – Conditional Probability
**What is done:** Probability of default when credit score is less than 600.

**Why:** Used to identify high-risk customers.

```
low=df[df['Credit_Score']<600]
cond=(low['Default_Status']=='Yes').mean()
print('Conditional Probability:',cond)
```

## Step 8 – Distribution Visualization
**What is done:** Histogram of Credit Score with KDE curve.

**Why:** Helps visualize distribution and check if data follows normal pattern.

```
sns.histplot(df['Credit_Score'],kde=True)
plt.title('Histogram of Credit Score')
plt.show()
```

## Step 9 – Skewness and Kurtosis
**What is done:** Skewness and Kurtosis of Loan Amount calculated.

**Why:** Helps identify shape of distribution (left/right skew and tails).

```
print('Skewness:',df['Loan_Amount'].skew())
print('Kurtosis:',df['Loan_Amount'].kurtosis())
```

## Step 10 – Q-Q Plot
**What is done:** Q-Q plot of Income created.

**Why:** Helps check whether data follows normal distribution.

```
stats.probplot(df['Income'],dist='norm',plot=plt)
plt.title('Q-Q Plot of Income')
plt.show()
```

## Step 11 – Linear Algebra Operations
**What is done:** Dot Product, Norm and Angle between vectors calculated.

**Why:** These operations help measure similarity between customers.

```
vectors=df[['Income','Loan_Amount']].head(5)
v1=vectors.iloc[0]
v2=vectors.iloc[1]

print('Dot Product:',np.dot(v1,v2))
print('Norm:',np.linalg.norm(v1))

cos=np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
angle=np.arccos(cos)
print('Angle:',angle)
```
