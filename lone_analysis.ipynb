# Part B Loan Analysis

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```
df=pd.read_csv('loan_applications.csv')
df.head()
```

```
print('Mean:',df['Income'].mean())
print('Median:',df['Income'].median())
print('Mode:',df['Income'].mode()[0])
```

```
loan_range=df['Loan_Amount'].max()-df['Loan_Amount'].min()
print('Range:',loan_range)
print('Variance:',df['Loan_Amount'].var())
print('Std Dev:',df['Loan_Amount'].std())
```

```
default_prob=(df['Default_Status']=='Yes').mean()
print('Default Probability:',default_prob)
```

```
df['Credit_Category']=pd.cut(df['Credit_Score'],bins=[300,600,700,850],labels=['Low','Medium','High'])
pd.crosstab(df['Default_Status'],df['Credit_Category'])
```

```
low=df[df['Credit_Score']<600]
cond=(low['Default_Status']=='Yes').mean()
print('Conditional Probability:',cond)
```

```
sns.histplot(df['Credit_Score'],kde=True)
plt.show()
```

```
print('Skewness:',df['Loan_Amount'].skew())
print('Kurtosis:',df['Loan_Amount'].kurtosis())
```

```
stats.probplot(df['Income'],dist='norm',plot=plt)
plt.show()
```

```
vectors=df[['Income','Loan_Amount']].head(5)
v1=vectors.iloc[0]
v2=vectors.iloc[1]

print('Dot:',np.dot(v1,v2))
print('Norm:',np.linalg.norm(v1))

cos=np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
angle=np.arccos(cos)
print('Angle:',angle)
```
